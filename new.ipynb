```python
def calculator():
    print("Welcome to the calculator!")
    print("You can enter operations like this: 12+5-3*2/4=")
    print("Available operations: + - * / % (modulus) and %p for percentage")

    user_input = input("Enter your operation: ").replace(" ", "")  
    print("your operation = " + user_input)
    
    if user_input.endswith("%p="):
        user_input = user_input.replace("%p=", "")
        numbers = user_input.split("/")
        if len(numbers) == 2:
            part = float(numbers[0])
            total = float(numbers[1])
            result = (part / total) * 100
            print("Percentage:", result, "%")
        else:
            print("Invalid percentage format. Use: part/total%p=")
        return

    
    if user_input.endswith("="):
        user_input = user_input[:-1]

    try:
        result = eval(user_input)
        print("Result:", result)
    except Exception as e:
        print("Invalid operation:", e)
calculator()
```
